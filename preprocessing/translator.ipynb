```python
from glob import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import joblib


expected_sales, the_number_of_shops = glob('expected_sales/*.csv'), glob('the_number_of_shops/*.csv')
expected_sales, the_number_of_shops
df1 = pd.concat(pd.read_csv(i) for i in expected_sales)
df2 = pd.concat(pd.read_csv(i) for i in the_number_of_shops)


df = df1.merge(
    df2,
    how = 'inner',
    on = ['기준_년분기_코드', '상권_구분_코드_명', '상권_코드_명', '서비스_업종_코드_명']
)

# 유지할 컬럼
keep_cols1 = [
    '상권_구분_코드_x',
    '상권_구분_코드_명',
]

keep_cols2 = [
    
    '상권_코드_x',
    '상권_코드_명',
    
]

keep_cols3 = [
    
    '서비스_업종_코드_x',
    '서비스_업종_코드_명'
]

# 1. 지정한 컬럼만 남기고 나머지 제거
df_filtered1 = df[keep_cols1]
df_filtered2 = df[keep_cols2]
df_filtered3 = df[keep_cols3]

# 2. 중복 제거
df_unique1 = df_filtered1.drop_duplicates()
df_unique2 = df_filtered2.drop_duplicates()
df_unique3 = df_filtered3.drop_duplicates()

# 코드 → 한글명 매핑 딕셔너리 생성
code_name_dicts = {
    '상권_구분_코드': dict(zip(df_unique1['상권_구분_코드_x'], df_unique1['상권_구분_코드_명'])),
    '상권_코드': dict(zip(df_unique2['상권_코드_x'], df_unique2['상권_코드_명'])),
    '서비스_업종_코드': dict(zip(df_unique3['서비스_업종_코드_x'], df_unique3['서비스_업종_코드_명']))
}

bundle = joblib.load('dict_trans.joblib')
```
